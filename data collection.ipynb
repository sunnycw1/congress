# 2. Presidential debates

get all the links from the menu to the left of 


'https://www.debates.org/index.php?page=debate-transcripts'

```python
url = 'https://www.debates.org/index.php?page=debate-transcripts'
header = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
r = requests.get(url) #, headers=header)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

use Chrome and 'inspect' to see the name to give to soup.find

```python
debates = soup.find("ul", {"class" : "arrow1"}).find_all("li")[9]
```

```python
debate_links = debates.find_all('a', href=True)
debate_urls = [link.get('href') for link in debate_links]
debate_urls = debate_urls[:-1] # last link not a debate
debate_urls
```

```python
df_debates = pd.DataFrame(columns=['title', 'text'])

for i, debate_url in enumerate(debate_urls):
    r = requests.get(debate_url)
    soup = BeautifulSoup(r.content, 'html.parser')
    content = soup.find("div", {"id": "content-sm"})
    
    title = content.find('h1').getText()
    print(title)

    text = content.getText()
    
    df_debates.loc[i] = [title, text]
```

```python
df_debates.head()
```

```python
df_debates.to_csv('presidential_debates.csv')
```
